## In this notebook, we will compare the performance of the models with different configurations for Value Embedding:
 - MLP
 - 1D Convolution
 - 1D Convolution + Pretrained

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
baseline = pd.read_csv('checkpoints/SMAP_BERT_f1_scores.csv')
pad_1d = pd.read_csv('checkpoints/SMAP_1DConv_f1_scores.csv')
pad = pd.read_csv('checkpoints/SMAP_f1_scores.csv')
```

```python
pad
```

```python
# Plot f1 scores through the steps
limit = 33
plt.figure(figsize=(6, 8))
plt.plot(baseline['f1_score'][:limit], label='MLP')
plt.plot(pad_1d['f1_score'][:limit], label='1D Convolution')
plt.plot(pad['f1_score'][:limit], label='1D Convolution + Pretrained')
plt.legend()
plt.xlabel('Step')
plt.ylabel('F1 Score')
plt.title('F1 Score Comparison for SMAP Dataset')
plt.show()
```
